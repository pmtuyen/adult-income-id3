## Xử lý dữ liệu

```python
import pandas as pd
import re
```

```python
df_train = pd.read_csv("/content/drive/MyDrive/dataset/data_train.csv")
df_val = pd.read_csv("/content/drive/MyDrive/dataset/data_val.csv")
df_test = pd.read_csv("/content/drive/MyDrive/dataset/data_test.csv")
```

```python
print("Train:")
print(df_train.info())
print("\nValidation:")
print(df_val.info())
print("\nTest:")
print(df_test.info())
```

```python
#Xử lý câu không liên quan
def clean_text(text):
    # Chuyển về str và xử lý NaN
    if pd.isna(text):
        return ""
    text = str(text).lower().strip()
    # Loại bỏ các chuỗi ký tự censore
    if re.match(r'^[a-z0-9\s]{10,}$', text) and len(set(text)) < 5:
        return ""
    # Loại bỏ bình luận "nhận xu"
    if "nhận xu" in text or "mang tính chất minh họa" in text:
        return ""
    return text
```

```python
# Áp dụng làm sạch cho cột 'data'
df_train['data'] = df_train['data'].apply(clean_text)
df_val['data'] = df_val['data'].apply(clean_text)
df_test['data'] = df_test['data'].apply(clean_text)

# Loại bỏ các dòng mà cột 'data' rỗng
df_train = df_train[df_train['data'] != ""]
df_val = df_val[df_val['data'] != ""]
df_test = df_test[df_test['data'] != ""]
```

```python
print("Train:")
print(df_train.info())
print("\nValidation:")
print(df_val.info())
print("\nTest:")
print(df_test.info())
```

```python
def normalize_text(text):
    # Loại bỏ ký tự đặc biệt và biểu tượng cảm xúc
    text = re.sub(r'[^\w\s]', '', text)
    # Tách từ tiếng Việt
    text = word_tokenize(text, format="text")
    return text
```

```python
df_train['data_clean'] = df_train['data'].apply(normalize_text)
df_val['data_clean'] = df_val['data'].apply(normalize_text)
df_test['data_clean'] = df_test['data'].apply(normalize_text)
```

```python
# Gán nhãn 'neutral' cho các giá trị NaN trong cột khía cạnh
aspect_columns = ['stayingpower', 'texture', 'smell', 'price', 'others', 'colour', 'shipping', 'packing']
for col in aspect_columns:
    df_train[col] = df_train[col].fillna('neutral')
    df_val[col] = df_val[col].fillna('neutral')
    df_test[col] = df_test[col].fillna('neutral')
```

```python
# Thống kê phân bố nhãn
for col in aspect_columns:
    print(f"\nNhãn cho {col}:")
    print(df_train[col].value_counts())
```

```python
# Tạo cột nhị phân cho phát hiện khía cạnh
# Mã hóa các col khía cạnh
for col in aspect_columns:
    df_train[f'{col}_mentioned'] = df_train[col].apply(lambda x: 1 if x in ['positive', 'negative', 'neutral'] else 0)
    df_val[f'{col}_mentioned'] = df_val[col].apply(lambda x: 1 if x in ['positive', 'negative', 'neutral'] else 0)
    df_test[f'{col}_mentioned'] = df_test[col].apply(lambda x: 1 if x in ['positive', 'negative', 'neutral'] else 0)
```

```python
# Mã hóa các col cảm xúc
label_map = {'positive': 2, 'negative': 0, 'neutral': 1}
for col in aspect_columns:
    df_train[f'{col}_label'] = df_train[col].map(label_map)
    df_val[f'{col}_label'] = df_val[col].map(label_map)
    df_test[f'{col}_label'] = df_test[col].map(label_map)
```

```python
# Save data clean
df_train.to_csv("/content/drive/MyDrive/dataset/data_train_clean.csv", index=False)
df_val.to_csv("/content/drive/MyDrive/dataset/data_val_clean.csv", index=False)
df_test.to_csv("/content/drive/MyDrive/dataset/data_test_clean.csv", index=False)
```

```python
def encode_text(text, max_length=256):
    inputs = tokenizer(text, return_tensors="pt", max_length=max_length, truncation=True, padding=True)
    return inputs
```

```python
df_train['colour_input'] = df_train['data_clean'].apply(lambda x: encode_text(x))
```

```python
class AspectDataset(torch.utils.data.Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=256):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        encoding = self.tokenizer(text, max_length=self.max_length, truncation=True, padding='max_length', return_tensors='pt')
        item = {key: val.squeeze() for key, val in encoding.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item
```

```python
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base")
```

```python
aspect_columns = ['stayingpower', 'texture', 'smell', 'price', 'others', 'colour', 'shipping', 'packing']
# Tạo datasets cho tất cả các khía cạnh
train_datasets = {}
val_datasets = {}
test_datasets = {}
```

```python
for aspect in aspect_columns:
    label_column = f'{aspect}_label'

    # Kiểm tra xem cột nhãn có tồn tại không
    if label_column not in df_train.columns or label_column not in df_val.columns or label_column not in df_test.columns:
        print(f"Warning: {label_column} not found in DataFrame. Skipping...")
        continue

    # Tạo dataset cho tập train
    train_datasets[aspect] = AspectDataset(
        texts=df_train['data_clean'].tolist(),
        labels=df_train[label_column].tolist(),
        tokenizer=tokenizer,
        max_length=256
    )

    # Tạo dataset cho tập validation
    val_datasets[aspect] = AspectDataset(
        texts=df_val['data_clean'].tolist(),
        labels=df_val[label_column].tolist(),
        tokenizer=tokenizer,
        max_length=256
    )

    # Tạo dataset cho tập test
    test_datasets[aspect] = AspectDataset(
        texts=df_test['data_clean'].tolist(),
        labels=df_test[label_column].tolist(),
        tokenizer=tokenizer,
        max_length=256
    )

    print(f"Created datasets for aspect: {aspect}")
    print(f"Train dataset size: {len(train_datasets[aspect])}")
    print(f"Validation dataset size: {len(val_datasets[aspect])}")
    print(f"Test dataset size: {len(test_datasets[aspect])}")
```

```python
# Hàm tính toán chỉ số đánh giá
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    return {
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
# Huấn luyện mô hình cho từng khía cạnh
for aspect in aspect_columns:
    print(f"\n=== Training model for aspect: {aspect} ===")

    # Khởi tạo mô hình
    model = AutoModelForSequenceClassification.from_pretrained("vinai/phobert-base", num_labels=3)

    # Đường dẫn lưu mô hình và kết quả
    output_dir = f"/content/drive/MyDrive/dataset/train.result/phobert_{aspect}"
    report_file = f"/content/drive/MyDrive/dataset/train.result/report_{aspect}.txt"

    # Cấu hình huấn luyện (sử dụng eval_strategy thay cho evaluation_strategy)
    training_args = TrainingArguments(
        output_dir=output_dir,
        num_train_epochs=3,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        eval_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        metric_for_best_model='f1',
        logging_dir=f"/content/drive/MyDrive/dataset/train.result/logs_{aspect}",
        logging_steps=10,
        save_total_limit=2,
        report_to="none"
    )

    # Khởi tạo Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_datasets[aspect],
        eval_dataset=val_datasets[aspect],
        compute_metrics=compute_metrics
    )

    # Huấn luyện mô hình
    trainer.train()

    # Đánh giá trên tập test
    print(f"\nEvaluating on test set for aspect: {aspect}")
    test_results = trainer.evaluate(test_datasets[aspect])
    print(f"Test Results for {aspect}: {test_results}")
```
